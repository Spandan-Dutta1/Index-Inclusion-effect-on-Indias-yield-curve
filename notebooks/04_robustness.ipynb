```python
#Robustness checks
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.iv import IV2SLS


dataset = pd.read_csv(r'C:\Users\Spandan Dutta\econometrics_project\data\processed\master_panel.csv',parse_dates=['date'], index_col='date')

# differenced columns
diff_cols = ['yield_2y', 'yield_5y', 'yield_10y', 'yield_30y','slope_10y_2y', 'slope_30y_10y','us_10y', 'vix', 'dxy', 'brent', 'laf_net']

for col in diff_cols:
    dataset[f'd_{col}'] = dataset[col].diff()
```

```python
#2SLS
controls = ['d_us_10y', 'd_vix', 'd_dxy', 'd_brent', 'repo_rate', 'cpi', 'd_laf_net']
X = ['d_yield_10y', 'fpi', 'jpmorgan_weight'] + controls
df_base = dataset[X].dropna()

baseline = IV2SLS(df_base['d_yield_10y'],sm.add_constant(df_base[controls]),df_base[['fpi']],instruments = df_base[['jpmorgan_weight']]).fit(cov_type='robust')
print(baseline.summary)
```

```python
#RC1: cpi_inflation instead of cpi 
controls_rc1 = ['d_us_10y', 'd_vix', 'd_dxy', 'd_brent', 'repo_rate', 'cpi_inflation', 'd_laf_net']
reg_vars_rc1 = ['d_yield_10y', 'fpi', 'jpmorgan_weight'] + controls_rc1
df_rc1 = dataset[reg_vars_rc1].dropna()
X = sm.add_constant(df_rc1[controls_rc1])
rc1 = IV2SLS(df_rc1['d_yield_10y'],X, df_rc1[['fpi']],instruments = df_rc1[['jpmorgan_weight']]).fit(cov_type='robust')
print(rc1.summary)
#fpi baseline = -0.1446 
#fpi rc1 =  -0.0054
#change =0.1392
```

```python
#robustness check 2
controls_rc2 = ['d_us_10y', 'd_vix', 'd_brent', 'repo_rate', 'cpi', 'd_laf_net']
reg_vars_rc2 = ['d_yield_10y', 'fpi', 'jpmorgan_weight'] + controls_rc2
df_rc2 = dataset[reg_vars_rc2].dropna()
X = sm.add_constant(df_rc2[controls_rc2])
rc2 = IV2SLS(df_rc2['d_yield_10y'],X,df_rc2[['fpi']],instruments = df_rc2[['jpmorgan_weight']]).fit(cov_type='robust')
print(rc2.summary)
#baseline :-0.1446  
#rc2_fpi :-0.1440
#change : 0.0006
```

```python
#robustness check 3: 2021 to 2022
controls_rc3 = ['d_us_10y', 'd_vix', 'd_dxy', 'd_brent', 'repo_rate', 'cpi', 'd_laf_net']
reg_vars_rc3 = ['d_yield_10y', 'fpi', 'jpmorgan_weight'] + controls_rc3
df_rc3 = dataset[reg_vars_rc3].dropna()
df_rc3 = df_rc3[df_rc3.index >= '2022-01-01']
X = sm.add_constant(df_rc3[controls_rc3])
rc3 = IV2SLS(df_rc3['d_yield_10y'],X,df_rc3[['fpi']],instruments = df_rc3[['jpmorgan_weight']]).fit(cov_type='robust')
print(rc3.summary)
#baseline : -0.1446 #rc3 :-0.1797  #change : -0.0351 
```
