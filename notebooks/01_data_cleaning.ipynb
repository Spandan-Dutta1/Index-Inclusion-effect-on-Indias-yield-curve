```python
#Government Securities
df1 =pd.read_excel(file['gsec'], header=None)
df1.head()
year_rows = {
    '2021-22': 131,
    '2022-23': 100,
    '2023-24': 69,
    '2024-25': 38,
    '2025-26': 7
}
months_list = ['Apr','May','Jun','Jul','Aug','Sep',
               'Oct','Nov','Dec','Jan','Feb','Mar']
col_indices  = list(range(2, 14))
month_num    = {
    'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,
    'Oct':10,'Nov':11,'Dec':12,'Jan':1,'Feb':2,'Mar':3
}
#extracting 2025-26
year2025_2026 = []
start_row = 7
base = 2025
for col_indx , month in zip(col_indices, months_list):
    year = base + 1 if month in ['Jan','Feb','Mar'] else base #jan feb mar in 2026
    #creating time stamp
    date = pd.Timestamp(year=year, month=month_num[month], day=28)
    row_data1 = {'date': date}
    for tenor in range(1, 31):
        val = df1.iloc[start_row + tenor, col_indx]
        try:
            val = float(val) if val != '-' else np.nan
        except:
            val = np.nan
        row_data1[f'yield_{tenor}y'] = val
    
    year2025_2026.append(row_data1)
test_df1 = pd.DataFrame(year2025_2026)

#extracting for 2024-25
year2024_2025 = []
start_row = 38
base = 2024
for col_indx , month in zip(col_indices, months_list):
    year = base + 1 if month in ['Jan','Feb','Mar'] else base #jan feb mar in 2025
    #creating time stamp
    date = pd.Timestamp(year=year, month=month_num[month], day=28)
    row_data2 = {'date': date}
    for tenor in range(1, 31):
        val = df1.iloc[start_row + tenor, col_indx]
        try:
            val = float(val) if val != '-' else np.nan
        except:
            val = np.nan
        row_data2[f'yield_{tenor}y'] = val
    
    year2024_2025.append(row_data2)
test_df2 = pd.DataFrame(year2024_2025)

#extracting 2023-24
year2023_2024 = []
start_row = 69
base = 2023
for col_indx , month in zip(col_indices, months_list):
    year = base + 1 if month in ['Jan','Feb','Mar'] else base 
    #creating time stamp
    date = pd.Timestamp(year=year, month=month_num[month], day=28)
    row_data3 = {'date': date}
    for tenor in range(1, 31):
        val = df1.iloc[start_row + tenor, col_indx]
        try:
            val = float(val) if val != '-' else np.nan
        except:
            val = np.nan
        row_data3[f'yield_{tenor}y'] = val
    
    year2023_2024.append(row_data3)
test_df3 = pd.DataFrame(year2023_2024)

#extracting 2022-23
year2022_2023 = []
start_row = 100
base = 2022
for col_indx , month in zip(col_indices, months_list):
    year = base + 1 if month in ['Jan','Feb','Mar'] else base 
    #creating time stamp
    date = pd.Timestamp(year=year, month=month_num[month], day=28)
    row_data4 = {'date': date}
    for tenor in range(1, 31):
        val = df1.iloc[start_row + tenor, col_indx]
        try:
            val = float(val) if val != '-' else np.nan
        except:
            val = np.nan
        row_data4[f'yield_{tenor}y'] = val
    
    year2022_2023.append(row_data4)
test_df4 = pd.DataFrame(year2022_2023)

#extracting 2021-22
year2021_2022 = []
start_row = 131
base = 2021
for col_indx , month in zip(col_indices, months_list):
    year = base + 1 if month in ['Jan','Feb','Mar'] else base 
    #creating time stamp
    date = pd.Timestamp(year=year, month=month_num[month], day=28)
    row_data5 = {'date': date}
    for tenor in range(1, 31):
        val = df1.iloc[start_row + tenor, col_indx]
        try:
            val = float(val) if val != '-' else np.nan
        except:
            val = np.nan
        row_data5[f'yield_{tenor}y'] = val
    
    year2021_2022.append(row_data5)
test_df5 = pd.DataFrame(year2021_2022)

final_testdf = pd.concat([test_df5, test_df4, test_df3,test_df2, test_df1],axis=0,ignore_index=True) 


final_testdf = final_testdf.sort_values('date').reset_index(drop=True)
final_testdf = final_testdf.dropna(subset=['yield_1y'])
final_testdf = final_testdf[final_testdf['date'] >= '2021-01-01'].reset_index(drop=True)
final_testdf['date'] = final_testdf['date'] + pd.offsets.MonthEnd(0)
final_df_gsec = final_testdf.set_index('date')
```

```python
#Foreign Portfolio Investment - Quarterly data
df2 = pd.read_excel(file['fpi'],sheet_name='Ownership Pattern of Government',header=None)
#renaming the columns
test2_df = df2.iloc[6:25, 1:13].copy()
test2_df.columns = [
    'date',            
    'total_crores',    
    'commercial_banks',
    'co_op_banks',      
    'non_bank_pds',    
    'insurance',       
    'mutual_funds',     
    'provident_funds', 
    'pension_funds',   
    'fin_institutions',
    'corporates',      
    'fpi'          
]

#fixing data type
test2_df["date"] = pd.to_datetime(test2_df["date"]) #converting it into date time format
test2_df["fpi"] = pd.to_numeric(test2_df["fpi"]) #converting it into numeric format
#filter from 2021
test2_df = test2_df[test2_df['date'] >= '2021-01-01'].sort_values('date')

#Linear Interpolation
#setting the date column as index 
test2_indexed = test2_df.set_index('date')[["fpi"]]
#filling with interpolation
final_df_fpi = test2_indexed.resample("ME").interpolate(method="linear")
```

```python
#US 10 Year
df4 = pd.read_csv(file['us10y'],parse_dates=['observation_date'],index_col='observation_date')
#replacing . with nan
df4 = df4.replace(".", np.nan)
df4 = df4.astype(float)
df4 = df4.rename(columns={'DGS10': 'us_10y'})
print(df4.head(5))
print(df4.isnull().sum().values[0])
#dropping nan and converting it into monthly by taking the month end values
df4_clean = df4.dropna()
final_df_us10y= df4_clean.resample("ME").last() 
```

```python
#US 2 Year
df5 = pd.read_csv(file['us2y'],parse_dates=['observation_date'],index_col='observation_date')
# Clean
df5 = df5.replace('.', np.nan)
df5 = df5.astype(float)
df5 = df5.rename(columns={'DGS2': 'us_2y'})
df5 = df5.dropna()

# Collapse to monthly
final_df_us2y = df5.resample('ME').last()
```

```python
#Dollar Index(DXY)
df6 = pd.read_csv(file['dxy'],parse_dates=['observation_date'],index_col='observation_date')
df6 = df6.replace('.', np.nan)
df6 = df6.astype(float)
df6 = df6.rename(columns={'DTWEXBGS': 'dxy'})
df6 = df6.dropna()
final_df_dxy = df6.resample('ME').last()

#It is a price variable so we can also consider its mean rather than month end values
df6_last = df6.resample('ME').last()
df6_mean = df6.resample('ME').mean()
```

```python
#VIX
df7 = pd.read_csv(file['vix'],parse_dates=['observation_date'],index_col='observation_date')

# Clean
df7 = df7.replace('.', np.nan)
df7 = df7.astype(float)
df7 = df7.rename(columns={'VIXCLS': 'vix'})
df7 = df7.dropna()

# Collapse to monthly
final_df_vix = df7.resample('ME').mean()

#for robustness check 
vix_last = df7.resample('ME').last()
```

```python
#Brent OIL 
df8 = pd.read_csv(file['brent'],parse_dates=['observation_date'],index_col='observation_date')

# Clean
df8 = df8.replace('.', np.nan)
df8 = df8.astype(float)
df8 = df8.rename(columns={'DCOILBRENTEU': 'brent'})
df8 = df8.dropna()

# Collapse to monthly
final_df_brent = df8.resample('ME').mean()

#for robustness check
brent_last = df8.resample('ME').last()
```

```python
#Indian CPI 
df9 = pd.read_csv(file['cpi'],parse_dates=['observation_date'],index_col='observation_date')

# Clean
df9 = df9.replace('.', np.nan)
df9 = df9.astype(float)
df9 = df9.rename(columns={'INDCPIALLMINMEI': 'cpi'})

# CPI is already monthly so no resampling needed alignment index to month end
df9.index = df9.index + pd.offsets.MonthEnd(0)

# Calculate inflation rate (year on year % change)
df9['cpi_inflation'] = df9['cpi'].pct_change(12) * 100
# Filtering from 2021
final_df_cpi = df9[df9.index >= '2021-01-31']
```

```python
#LAF
df10 = pd.read_excel(file['laf'],sheet_name='Report 1',header=None)
df10_sub = df10.iloc[8:, [1, 18]].copy()
df10_sub.columns = ['date', 'laf_net']


df10_sub['date']    = pd.to_datetime(df10_sub['date'], errors='coerce')
df10_sub['laf_net'] = pd.to_numeric(df10_sub['laf_net'], errors='coerce')
df10_sub = df10_sub.dropna(subset=['date'])
df10_sub= df10_sub[df10_sub['date'] >= '2021-01-01'].sort_values('date')

# monthly average
final_df_laf = (df10_sub.set_index('date')[['laf_net']].resample('ME').mean())
```

```python
df11 = pd.read_excel(r"C:\Users\Spandan Dutta\econometrics_project\data\temp\Indian_Repo_Rate_Data.xlsx")
df11['Dictionary Date'] = (df11['Dictionary Date'].astype(str).str.replace("'", "", regex=False).str.strip())

df11['Dictionary Date'] = pd.to_datetime(df11['Dictionary Date'])
df11= df11.sort_values('Dictionary Date')
#month end index
date_range = pd.date_range(start='2021-01-31', end='2026-03-31', freq='ME')
final_df_repo = pd.DataFrame(index=date_range, columns=['repo_rate'], dtype=float)
final_df_repo['repo_rate'] = 4.00
# Applying each rate change forward from that month-end onward
for _, row in df11.iterrows():
    change_date = row['Dictionary Date'] + pd.offsets.MonthEnd(0)  # safe month-end alignment
    new_rate = float(row['Repo Rate'])
    final_df_repo.loc[change_date:, 'repo_rate'] = new_rate
```

```python
#JP morgan weights
jpmorgan_schedule = {
    '2024-06-30': 1.0,
    '2024-07-31': 2.0,
    '2024-08-31': 3.0,
    '2024-09-30': 4.0,
    '2024-10-31': 5.0,
    '2024-11-30': 6.0,
    '2024-12-31': 7.0,
    '2025-01-31': 8.0,
    '2025-02-28': 9.0,
    '2025-03-31': 10.0
}

#monthly series
date_range = pd.date_range(start='2021-01-31',end='2026-03-31',freq='ME')
final_df_jpm = pd.DataFrame(index=date_range,columns=['jpmorgan_weight'],dtype=float)
# Pre-inclusion weight = 0
final_df_jpm['jpmorgan_weight'] = 0.0

# Fill in phase-in weights
for date_str, weight in jpmorgan_schedule.items():
    final_df_jpm.loc[pd.Timestamp(date_str):, 'jpmorgan_weight'] = weight
```

```python
#master panel
dataframes = {
    'yields'     : final_df_gsec,
    'fpi'        : final_df_fpi,
    'us10y'      : final_df_us10y,
    'us2y'       : final_df_us2y,
    'dxy'        : final_df_dxy,
    'vix'        : final_df_vix,
    'brent'      : final_df_brent,
    'cpi'        : final_df_cpi,
    'laf'        : final_df_laf,
    'repo'       : final_df_repo,
    'jpm'        : final_df_jpm,
}

master = final_df_gsec[['yield_1y','yield_2y','yield_5y',
                          'yield_10y','yield_30y']].copy()

# Join all datasets
master = master.join(final_df_fpi,   how='left')
master = master.join(final_df_us10y, how='left')
master = master.join(final_df_us2y,  how='left')
master = master.join(final_df_dxy,   how='left')
master = master.join(final_df_vix,   how='left')
master = master.join(final_df_brent, how='left')
master = master.join(final_df_cpi,   how='left')
master = master.join(final_df_laf,   how='left')
master = master.join(final_df_repo,  how='left')
master = master.join(final_df_jpm,   how='left')

# Add derived variables
master['post_announcement'] = (master.index >= '2023-09-30').astype(int)
master['post_inclusion']    = (master.index >= '2024-06-30').astype(int)
master['slope_10y_2y']      = master['yield_10y'] - master['yield_2y']
master['slope_30y_10y']     = master['yield_30y'] - master['yield_10y']
master['slope_5y_2y']       = master['yield_5y']  - master['yield_2y']
master['us_slope']          = master['us_10y']     - master['us_2y']

# First differences
diff_cols = ['yield_1y','yield_2y','yield_5y','yield_10y',
             'yield_30y','slope_10y_2y','fpi','us_10y',
             'vix','brent','dxy','cpi_inflation',
             'laf_net','repo_rate']
for col in diff_cols:
    master[f'd_{col}'] = master[col].diff()
```

```python
# Reset index ONCE — date becomes a column
master = master.reset_index()
master = master.rename(columns={'date': 'date'})

```
